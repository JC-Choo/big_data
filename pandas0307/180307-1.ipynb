## 여러 개의 데이터를 하나로 합치기

```python
transaction2017 = pd.read_csv("transaction.csv")
transaction2017.head()
```

```python
transaction2016 = pd.read_csv("transaction2016.csv")
transaction2016.head()
```

```python
transaction2015 = pd.read_csv("transaction2015.csv")
transaction2015.head()
```

```python
# concat == concatenate, colums이 동일해야 함
# 1. 기준 열(key column)을 사용하지 않고 단순히 데이터를 연결(concatenate)한다.
# 2. 기본적으로는 위/아래로 데이터 행을 연결한다. 단순히 두 시리즈나 데이터프레임을 연결하기 때문에 인덱스 값이 중복될 수 있다.


pd.concat([transaction2015, transaction2016, transaction2017], axis=0) #리스트로 여러개의 데이터를 넣을 수 있음
```

```python
transaction2015.append([transaction2016, transaction2017])
```

```python
transaction2015 = transaction2015.set_index("Name")
transaction2015
```

```python
# profile = [
#     {"Name" : "Kang", "phone" : "010-1234-5678"},
#     {"Name" : "Jang", "phone" : "010-5678-8765"},
#     {"Name" : "Kim", "phone" : "010-1234-8765"},
# ]
profile = [
    {"date" : "2017-01-01", "phone" : "010-1234-5678"},
    {"date" : "2017-01-05", "phone" : "010-5678-8765"},
    {"date" : "2017-01-09", "phone" : "010-1234-8765"},
]

profile = pd.DataFrame(profile)
profile
```

```python
#transaction 과 profile 을 합침 -> concat, merge 
# merge
# 1. 두 데이터 프레임의 공통 열 혹은 인덱스를 기준으로 두 개의 테이블을 합친다. 이 때 기준이 되는 열의 데이터를 키(key)라고 한다.
# 2. 공통 열을 기준으로 데이터를 찾아서 합친다. 기본적으로 양쪽 데이터프레임에 모두 키가 존재하는 데이터만 보여주는 inner join 방식을 사용.

# pd.merge(transaction2017, profile, how="inner") # kang and kim 만 겹치니까 2개만 나옴
# pd.merge(transaction2017, profile, how="outer") # outer : 안겹치는 부분도 살림
# pd.merge(transaction2017, profile, how="left") # left : 왼쪽 transaction2017을 기준으로 살림
# pd.merge(transaction2017, profile, how="right") # right : 오른쪽 profile을 기준으로 살림
pd.merge(transaction2017, profile)

# 여러개 하려면 merge 한 것을 변수로 가져와서 다시 하면 된다.
# temp = pd.merge(transaction2017, profile, how="left")
# pd.merge(temp, transaction2016)
```
