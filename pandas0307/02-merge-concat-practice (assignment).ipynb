** 10. 9번 데이터에서 현재 체중과 측정 체중이 가장 차이가 크게 나는 순으로 정렬해주세요. **

```python
# weightDataAll.dtypes
```

```python
def int_str(kg) :
    if pd.isnull(kg) :
        return kg
        
    if "kg" in kg :
        kg = kg.replace("kg", "")
    
    return float(kg)

weightDataAll["현재 체중"] = weightDataAll["현재 체중"].apply(int_str)
weightDataAll["목표 체중"] = weightDataAll["목표 체중"].apply(int_str)

# weightDataAll.head()
```

```python
# Write your code here!
weightDataAll["현재-측정 체중"] = weightDataAll["현재 체중"] - weightDataAll["측정 체중"]
weightDataAll["현재-측정 체중 abs"] = abs(weightDataAll["현재-측정 체중"])

weightDataAll.sort_values(by="현재-측정 체중 abs", ascending=False).head()
```
